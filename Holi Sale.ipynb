```python
# to import libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# to import CVS file

df = pd.read_csv(r'C:\Users\keshav\OneDrive\Desktop\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
# To Show the number of rows and column

df.shape
```

```python
# To show the top 10 rows

df.head(10)
```

```python
# Now we need to clean the data
```

```python
df.info()
```

```python
# for delete the unwanted rows
# axis = 1 is means full column

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# to check how many null value is available

pd.isnull(df).sum()
```

```python
# inplace used for save the Dataframe

df.dropna(inplace=True)
```

```python
df.shape
```

```python
# Change Datatype

df['Amount'] = df['Amount'].astype ('int')
```

```python
# Check DataType

df['Amount'].dtypes
```

```python
# check column 

df.columns
```

```python
# change column name 

df.rename(columns= {'Occupation':'Designation'})
```

```python
# check mean, min, max, count etc....

df.describe()
```

```python
# check specific mean, min, max, count etc....

df[['Age', 'Orders','Amount']].describe()
```

```python
#  Exploratory Data Analysis

df.columns
```

```python
sns.countplot(x= 'Gender', data = df) 
```

```python
ax = sns.countplot(x = 'Gender', data = df)

# count gender

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# women's have spent more money than men's

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

```python
# AGE

df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Most of the buyers are of age group between 26-35 yrs female

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Age Group', y= 'Amount', data = sales_age)
```

```python
# State 

df.columns
```

```python
# Higher order from Top 10 States

sales_State = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})

sns.barplot(x= 'State', y= 'Orders', data = sales_State)
```

```python
# Higher Amount spent from Top 10 States

sales_State = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})

sns.barplot(x= 'State', y= 'Amount', data = sales_State)
```

```python
# martial_status

df.columns
```

```python
ax = sns.countplot(data=df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# most of the buyer are married women's


sales_State = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(data = sales_State, x= 'Marital_Status', y= 'Amount', hue= 'Gender')
```

```python
# Occupation

df.columns
```

```python
ax = sns.countplot(data = df , x = 'Occupation')
sns.set(rc={'figure.figsize':(20,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Most of the Buyer are working in IT Sector, HealthCare  and Aviation


sales = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,7)})

sns.barplot(x= 'Occupation', y= 'Amount', data = sales)
```

```python
# Product Category 

df.columns
```

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set(rc={'figure.figsize':(25,7)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
# Most of the product are from Food, Clothing& Apparel and Electronics and Gadgets

prd_sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})

sns.barplot(x= 'Product_Category', y= 'Amount', data = prd_sales)
```

```python
# Top 10 Most sold Product

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
